```python
password=input('請輸入設定密碼(例如:一年十九班50號,密碼即:11950')
times=1
password2=input('請輸入密碼:')
while password !=password2 and times < 3:
  print('account')
  print('密碼錯誤!')
  times=times+1
  password2=input('請輸入密碼:')
if password ==password2:
  print('歡迎使用本系統')
else:
  print('輸入密碼超過三次,帳號已被鎖定')
number=input('請輸入座號:')

from random import randint
n=37
box=[]
for i in range(1,1+n):
  position=randint(0,i-1)
  box.insert(position,i)
print(box)
print('第一位被抽到的是:'+str(box[0])+'號')
if number ==str(box[0]):
  print('你是第一個被抽到')
else:
  print('你未被抽到')
```
